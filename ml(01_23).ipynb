```python
from shapely.geometry import Point, Polygon
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point, Polygon, LineString
import matplotlib.pyplot as plt
import matplotlib
import matplotlib.font_manager as fm
from geoband.API import *
import numpy as np
from tqdm import tqdm
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score
```

```python
## 어린이보호구역을 어린이교통사고 격자에 새로운 컬럼으로 생성


# 데이터 로딩
어린이보호구역 = pd.read_csv('data/9.오산시_어린이보호구역.csv')
어린이교통사고 = gpd.read_file('data/2.오산시_어린이교통사고_격자.geojson')

# point만들기
어린이보호구역['geometry'] = 어린이보호구역.apply(lambda row : Point([row['보호구역_경도'], row['보호구역_위도']]), axis=1)


# 보호구역 유무에 관한 새로운 컬럼 생성
어린이교통사고['어린이보호구역'] = 0

어린이보호구역 = gpd.GeoDataFrame(어린이보호구역, geometry='geometry')
buf_poly = gpd.GeoDataFrame({'geometry': 어린이보호구역.buffer(0.003)})
어린이보호구역['geometry'] = buf_poly
어린이보호구역

for a in range(0, len(어린이보호구역)):
    for i in range(0, len(어린이교통사고)):
        if (어린이보호구역.geometry[a].intersects(어린이교통사고.geometry[i])) == False :
            pass
        else :
            어린이교통사고['어린이보호구역'][i] += 1
            
            
어린이보호구역 = 어린이교통사고['어린이보호구역']
어린이보호구역


어린이교통사고 = pd.read_csv('filename1.csv')


어린이교통사고['어린이보호구역'] = 어린이보호구역
어린이교통사고
```

# 머신러닝 관련 알고리즘

```python
from xgboost import plot_importance # 중요변수 시각화
from sklearn.datasets import load_boston # dataset
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error,mean_absolute_error
from xgboost import XGBRegressor # 회귀트리 모델
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LassoCV
from sklearn.linear_model import ElasticNetCV
from sklearn.svm import LinearSVR
from sklearn.linear_model import BayesianRidge
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import GradientBoostingRegressor
from lightgbm import LGBMRegressor
```

```python
X =df_i2.drop(['accident_cnt'],axis=1)
y=df_i2['accident_cnt']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
```

```python
pd.options.display.float_format = '{:.10f}'.format
```

```python
# linear
linear = LinearRegression()
cross = cross_val_score(linear,X,y,cv=5,scoring='neg_mean_absolute_error')
print('linear(cross-val-score.mean):{:.3f}'.format(-cross.mean()))
```

```python
# RandomForest
rf = RandomForestRegressor()
cross = cross_val_score(rf,X,y,cv=5,scoring='neg_mean_absolute_error')
print('rf(cross-val-score.mean):{:.3f}'.format(-cross.mean()))
```

```python
# LassoCV
lasso = LassoCV()
cross = cross_val_score(lasso,X,y,cv=5,scoring='neg_mean_absolute_error')
print('lasso(cross-val-score.mean):{:.3f}'.format(-cross.mean()))
```

```python
# ElasticNetCV
elastic = ElasticNetCV()
cross = cross_val_score(elastic,X,y,cv=5,scoring='neg_mean_absolute_error')
print('elastic(cross-val-score.mean):{:.3f}'.format(-cross.mean()))
```

```python
# LinearSVR
svr = LinearSVR()
cross = cross_val_score(svr,X,y,cv=5,scoring='neg_mean_absolute_error')
print('svr(cross-val-score.mean):{:.3f}'.format(-cross.mean()))
```

```python
# BayesianRidge
bayesian = BayesianRidge()
cross = cross_val_score(bayesian,X,y,cv=5,scoring='neg_mean_absolute_error')
print('bayesian(cross-val-score.mean):{:.3f}'.format(-cross.mean()))
```

```python
# KNeighbors
knn = KNeighborsRegressor()
cross = cross_val_score(knn,X,y,cv=5,scoring='neg_mean_absolute_error')
print('knn(cross-val-score.mean):{:.3f}'.format(-cross.mean()))
```

```python
# DecisionTree
decisiontree = DecisionTreeRegressor()
cross = cross_val_score(decisiontree,X,y,cv=5,scoring='neg_mean_absolute_error')
print('decisiontree(cross-val-score.mean):{:.3f}'.format(-cross.mean()))
```

```python
# LGBM
lgbm = LGBMRegressor()
cross = cross_val_score(lgbm,X,y,cv=5,scoring='neg_mean_absolute_error')
print('lgbm(cross-val-score.mean):{:.3f}'.format(-cross.mean()))
```

```python
# XGB
xg = XGBRegressor()
cross = cross_val_score(xg,X,y,cv=5,scoring='neg_mean_absolute_error')
print('xg(cross-val-score.mean):{:.3f}'.format(-cross.mean()))
```

```python
# GradientBoosting
gradient = GradientBoostingRegressor()
cross = cross_val_score(gradient,X,y,cv=5,scoring='neg_mean_absolute_error')
print('gradient(cross-val-score.mean):{:.3f}'.format(-cross.mean()))
```

```python
# voting

from sklearn.ensemble import VotingClassifier, VotingRegressor
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_validate

# ensemble 할 model 정의
models = [
    ('svr', LinearSVR()),
    ('knn', KNeighborsRegressor()),
    ('gbr',GradientBoostingRegressor()),
    ('rf', RandomForestRegressor())
]
voting_regressor = VotingRegressor(models, n_jobs=-1) #n_jobs 코어수

cross = cross_val_score(voting_regressor,X,y,cv=5,scoring='neg_mean_absolute_error')
print('cross-val-score.mean \n{:.3f}'.format(-cross.mean()))
```

```python
# bagging
from sklearn.ensemble import BaggingRegressor
from sklearn.datasets import make_regression
regr = BaggingRegressor(base_estimator=LinearSVR(),
                        n_estimators=10, random_state=0).fit(X_train, y_train)
cross = cross_val_score(regr,X,y,cv=5,scoring='neg_mean_absolute_error')
print('cross-val-score.mean \n{:.3f}'.format(-cross.mean()))
```

```python
# stacking
from sklearn.ensemble import StackingRegressor
models = [
    ('svr', LinearSVR()),
    ('knn', KNeighborsRegressor()),
    ('gbr',GradientBoostingRegressor()),
    ('rf', RandomForestRegressor())
]
svr = LinearSVR()
stack_reg = StackingRegressor(models, final_estimator=svr, n_jobs=-1)
cross = cross_val_score(stack_reg,X,y,cv=5,scoring='neg_mean_absolute_error')
print('cross-val-score.mean \n{:.3f}'.format(-cross.mean()))
```
