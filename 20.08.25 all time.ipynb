```
merge_data=[]

for i in range(0,len(start_time)): # start time 길이면 될 듯
    if start_time[i][0] == '2020-01-01':
         break
    day_data=[]
    day_data.append([start_time[i][0],start_time[i][1]+'~'+end_time[i][1]])
    for j in range(0,1440):
        if exposure[i]+1 == len(day_data):
            break
        if int(start_time[i][1].split(":")[0]) == int(df_time['시간대'][j].split(":")[0]):
            if int(end_time[i][1].split(":")[1]) == 0:
                if int(start_time[i][1].split(":")[1]) <= int(df_time['시간대'][j].split(":")[1]) < 60:
                                day_data.append( [str(df_time['시간대'][j]),df_time[start_time[i][0]][j] ])
            else:
                if int(start_time[i][1].split(":")[1]) <= int(df_time['시간대'][j].split(":")[1]) < int(end_time[i][1].split(":")[1]):
                                day_data.append( [str(df_time['시간대'][j]),df_time[start_time[i][0]][j] ])
    else:
        merge_data.append(day_data)
```
